### Make Edges

```julia
start_node = Array{Int,1}(undef,0)
end_node = Array{Int,1}(undef,0)
c = Array{Float64}(undef,0)
```

```julia
ordered_key = sort(collect(keys(dic)))
for i in ordered_key
  for (node,v) in dic[i]
    push!(start_node,i+1)
    push!(end_node,node+1)
    push!(c,v)
  end
end
```

```julia
[start_node end_node]
```

```julia
minimum(start_node),maximum(end_node)
```

### Make Edges Dictionary

```julia
link_dict = OrderedDict{Tuple,Int64}()
for (i,link) in enumerate(zip(start_node,end_node))
  link_dict[link] = i
end
```

```julia
link_dict
```

```julia
origin = minimum(start_node)
destination = maximum(end_node)
```

### Number of nodes and number of links

```julia
no_node = max(maximum.([start_node,end_node])...) 
```

```julia
no_link = length(link_dict)
```

### Creating a graph

```julia
graph = Graph(no_node)
distmx = Inf*ones(no_node,no_node)
;
```

### Adding links to the graph

```julia
for (k,v) in link_dict
  add_edge!(graph,k)
  distmx[k...] = c[v]
end

# use graph's edges information
# do not use link_dict information
# update link_dict edge number with graph's
for (e_idx, e) in enumerate(edges(graph))
  link_dict[(src(e),dst(e))] = e_idx
end
```

### Run Dijkstra's Algorithm from the origin node to all nodes

```julia
state = dijkstra_shortest_paths(graph,origin,distmx)
```

### Retrieving the shortest path

```julia
path = enumerate_paths(state, destination)
println("The shortest path is:", path)
println("The shortest path is:", path .- 1)
```

### The cost of shortest path

```julia
println("Cost is $(state.dists[destination])") # directly from state
```
